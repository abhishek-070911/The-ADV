```python
import numpy as np
import pandas as pd
import random
```

```python
random.seed(44)

rows = []
object_types = ['Car','Motorcycle','Bike','Pedestrian','Truck','Animal']
speed_ranges = {
                'Car':(1.224,100),'Motorcycle':(1.224,100),
                'Bike':(1.44,30),'Pedestrian':(1.44,10),
                'Truck':(1.224,80),'Animal':(1.44,10)
               }

for i in range(1,10001):
    obj = random.choice(object_types)
    speed = random.uniform(*speed_ranges[obj])
    ttc = round(random.uniform(0.5, 8.0),2)
    braking = 1 if ttc <= 4 else 0
    avoided = 1 if braking == 1 and ttc >= 2 else 0

    rows.append({
        'Scenario_id' : i,
        'Object_type' : obj,
        'Object_speed_kmph' : speed,
        'time_to_collision_sec' : ttc,
        'braking_triggered' : braking,
        'collision_avoided' : avoided
    })

adv_df = pd.DataFrame(rows)
```
